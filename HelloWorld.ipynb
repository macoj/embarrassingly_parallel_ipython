```python
import ipyparallel as ipp
```

```python
client = ipp.Client()
```

```python
client.ids
```

```python
dview = client[:] 
```

```python
dview.block = True
```

## Pushing and pulling data

```python
dview.push(dict(a=3.14, b=2.89))
```

```python
dview.pull(('a', 'b'))
```

```python
def summing():
    return a + b
```

```python
dview.apply(summing)
```

Remote function

```python
@dview.remote(block=True)
def remote_function():
    import numpy as np
    return np.exp(10)
```

```python
remote_function()
```

## Scatter and gather

```python
dview.scatter('a', range(50))
dview.scatter('b', range(50))
```

```python
dview['a']
```

```python
def summing():
    import numpy as np
    return np.array(a) + np.array(b)
```

```python
dview.apply(summing)
```
